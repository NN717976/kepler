```julia
x = 0.50
y = 0
v_x = 0
v_y = 1.630
r = (x^2+y^2)^(1/2)
```

```julia
a_x = -x/r^3
a_y = -y/r^3
```

```julia
e = 0.1
v_x += e*a_x
v_y += e*a_y
```

```julia
x += e*v_x
y += e*v_y
```

```julia
list_x = []
list_y = []
x = 0.50
y = 0
v_x = 0
v_y = 1.63
r = (x^2+y^2)^(1/2)
e = 0.01
for _ in 1:200
    a_x = -x/r^3
    a_y = -y/r^3
    v_x += e*a_x
    v_y += e*a_y
    x += e*v_x
    y += e*v_y
    #println("x = ",x,",y = ",y)
    #println("v_x = ",v_x,",v_y = ",v_y)
    #println("a_x = ",a_x,",a_y = ",a_y)
    push!(list_x,x)
    push!(list_y,y)
end
plot(list_x,list_y)
```
